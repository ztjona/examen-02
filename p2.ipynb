```python
import numpy as np

# f = lambda x: (x - 1) ** 3 * (x + 2) * (x - 5)
f = lambda x: x**5 - 6 * x**4 + 2 * x**3 + 20 * x**2 - 27 * x + 10

import matplotlib.pyplot as plt

x = np.linspace(-4, 6, 100)
y = f(x)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x)")
plt.grid(True)
plt.ylim(-170, 60)
plt.show()
```

```python
from scipy import optimize

der = lambda x: 5 * x**4 - 24 * x**3 + 6 * x**2 + 40 * x - 27
x0 = -5
x1 = optimize.newton(f, x0, fprime=der)
print(x1)
x0 = 2
x2 = optimize.newton(f, x0, fprime=der)
print(x2)

x0 = 6
x3 = optimize.newton(f, x0, fprime=der)
print(x3)
```
